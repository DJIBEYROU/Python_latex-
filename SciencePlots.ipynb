```python
import matplotlib.pyplot as plt
import numpy as np

## Please select the style you want by uncommenting
#plt.style.use('science')
#plt.style.use(['science', 'grid'])
#plt.style.use(['science', 'notebook'])
#plt.style.use(['science', 'ieee'])
## I am using Nature journal figure 
plt.style.use(['science', 'nature'])
#plt.style.use(['science','no-latex'])
def sigmoid(x, a):
    return 1 / (1 + np.exp(-a * x))
x = np.linspace(-10,10,100)
y = sigmoid(x, 0.5)
y2 = sigmoid(x, 1)
y3 = sigmoid(x, 4)

# Figure
fig = plt.figure()
ax = fig.add_subplot(1,1,1) 
ax.plot(x,y,label = '$a = 0.5$')
ax.plot(x,y2,label = '$a = 1.0$')
ax.plot(x,y3,label = '$a = 4.0$')
ax.legend(title = 'Sigmoid Function',loc = 'lower right', fancybox = False, frameon = True) # 凡例の設定
ax.set(xlabel = '$x$')
ax.set(ylabel = '$y$')
plt.savefig('sigmoid.pdf')
```
